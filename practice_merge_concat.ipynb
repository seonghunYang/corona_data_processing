```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'customer_id': [1, 2, 3],
    'customer_name': ['Robert', 'Peter', 'Dave']
}, columns=['id', 'customer_id', 'customer_name'])
df1
```

```python
df2 = pd.DataFrame({
    'id': [1, 2, 4],
    'order_id': [100, 200, 300],
    'order_date': ['2021-01-21', '2021-02-03', '2020-10-01']
}, columns=['id', 'order_id', 'order_date'])
df2
```

```python
# 데이터 합치는 방법 
# 1. concat 두 데이터를 이어주는 방법 오왼, 위아래 그냥 붙임
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1)
```

```python
# 2. merge
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='id')
```

```python
# merge는 sql처럼 4가지의 결합 방법이 존재
pd.merge(df1, df2, how='inner')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
# 인덱스 기준으로 병합
df1 = df1.set_index('id')
```

```python
df2 = df2.set_index('id')
```

```python
pd.merge(df1, df2, left_index=True, right_index=True)
```

```python
pd.merge(df1, df2, left_index=True, right_index=True, how='outer')
```
