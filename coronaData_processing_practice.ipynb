```python
# 없는 데이터 처리하기 (결측치)
doc = pd.read_csv(PATH + "01-22-2020.csv", encoding='utf-8-sig')
doc.isnull().sum()
```

```python
# dropna 결측치 가진 행을 모두 삭제
doc = doc.dropna()
doc.head()
```

```python
doc = doc.dropna(subset=['Confirmed'])
```

```python
doc.head()
```

```python
# fillna 특정값으로 결측치를 대체
doc = doc.fillna(0)
doc.head()
```

```python
nan_data = {'Deaths': 0, 'Recovered': 0}
doc = doc.fillna(nan_data)
doc.head()
```
